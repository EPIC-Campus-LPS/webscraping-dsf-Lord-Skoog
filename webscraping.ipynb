### Link Scraper

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.spam.com'
html = requests.get(url)
s = BeautifulSoup(html.content, 'html.parser')

urls = []
finalurls = []
for link in s.find_all('a'):
    urls.append(link.get('href'))

for line in urls:
    if url in line:
        pass
    else:
        finalurls.append((line))
for line in finalurls:
    if line[0] == '/' or line[0] == '#':
        pass
    elif 'spam.com' in line:
        pass
    else:
        print(line)
```

### Weather

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

url = 'https://weather.com/weather/tenday/l/Littleton+CO?canonicalCityId=fdf4c3d397c75000204ff2ba1b9e9fa4adb284a0bf750f25ad6cd43c96cf9090'
html = requests.get(url)
s = BeautifulSoup(html.content, 'html.parser')

todayhigh = s.find_all('span', class_="DailyContent--temp--1s3a7")
i = 0
presentday = datetime.now()
print('             High:    Low:')
while i != 3:
    day = presentday + timedelta(i)
    print((day.strftime('%d-%m-%Y:  ')) +todayhigh[i].text +'      ' + (todayhigh[i+1].text))
    i += 1
```

### Recipe

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.epicurious.com/recipes/food/views/fresh-spring-rolls'
html = requests.get(url)
s = BeautifulSoup(html.content, 'html.parser')

a = s.find_all('div', class_="BaseWrap-sc-gjQpdd BaseText-ewhhUZ Description-cSrMCf iUEiRd ioVvSX fsKnGI")
print("Ingrediants: ")
for i in a:
    print(i.text)
b = s.find_all('ol', class_="InstructionGroupWrapper-bqiIwp ccobUj")
steps = []
for e in b:
    steps.append(e.text)
c = ''
d = c.join(steps)
d = d.split('.')
for i in d:
    print(i +'\n')
```
